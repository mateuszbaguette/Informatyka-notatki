# T: Programowanie zachłanne - problem pakowania plecaka

1. **Algorytm zachłanny (ang. greedy algorithm)** jest algorytmem, który w celu otrzymania rozwiązania zadanego problemu na każdym kroku zawsze dokonuje wyboru uznanego przez siebie w danej chwili za najkorzystniejszy. Wybór lokalnie optymalnej możliwości ma przynieść globalnie optymalne rozwiązanie.
2. **Optymalizacja** oznacza poszukiwanie rozwiązania zadanego zagadnienia, które zgodnie z ustalonym kryterium jest oceniane jako najlepsze z dopuszczalnych rozwiązań i równocześnie uwzględnia określone ograniczenia.

3. Algorytmy zachłanne służą do rozwiązywania problemów optymalizacyjnych. Problemy te zwykle można dość łatwo zdefiniować, jednakże nie istnieje prosty i szybki sposób ich rozwiązywania, który zawsze gwarantowałby osiągnięcie najlepszego możliwego wyniku.

4. Zasada działania algorytmów zachłannych jest następująca:
- w każdym kroku dokonują wyboru, który w tym konkretnym momencie wydaje się najlepszy. Innymi słowy, aby uzyskać najkorzystniejsze rozwiązanie całościowe, na każdym etapie działania wybierają to rozwiązanie, które lokalnie wydaje się najlepsze.
- idea polega na tym, aby przez dokonywanie zachłannych (czyli najlepszych w danym momencie) wyborów doprowadzić do optymalnego rozwiązania całego problemu.
- niestety nie zawsze udaje się w ten sposób uzyskać optymalne wyniki, ale w wielu przypadkach okazują się one wystarczające. Algorytmy zachłanne dają dobre efekty dla rozwiązań wielu problemów, kiedy istnieje wartość, którą trzeba maksymalizować lub minimalizować.

5. Programowanie zachłanne nie nadaje się do rozwiązania każdego problemu programistycznego. Przykładem sytuacji, kiedy algorytm zachłanny to dobry wybór, jest choćby często przywoływany w literaturze **„problem wyboru zajęć”**, zwany też „problemem planu zajęć. Zadanie to polega na przydzieleniu sal lekcyjnych do wszystkich odbywających się w szkole zajęć. Konieczne jest uwzględnienie ograniczeń: w danej sali w danym momencie mogą odbywać się tylko jedne zajęcia, a każde zajęcia mają swój czas rozpoczęcia i zakończenia. Celem jest odnalezienie możliwie największego podzbioru par pasujących do siebie sal i zajęć. Działanie algorytmu polega zatem na przyporządkowaniu konkretnych zajęć do sali, która w czasie, kiedy zaplanowano zajęcia, jest wolna. Operacja ta jest powtarzana do momentu, w którym wszystkie (o ile to możliwe) zaplanowane zajęcia zostaną przypisane do sal.

6. Przykładem problemu, do którego rozwiązania algorytmy zachłanne się nie nadają,może być **„problem kolorowania grafu”**. Kolorowanie grafu jest wykorzystywane w przypadku wielu złożonych problemów optymalizacyjnych (takich jak wskazanie możliwie najmniejszej liczby kolorów do pokolorowania mapy geograficznej). Celem działania algorytmu jest przypisanie wszystkim wierzchołkom (kolorowanie wierzchołkowe) lub wszystkim krawędziom (kolorowanie krawędziowe) danego grafu kolorów w taki sposób, aby żadne dwa sąsiadujące wierzchołki (sąsiadujące krawędzie) nie miały tego samego koloru. Optymalizacja w tym przypadku polega na wykorzystaniu możliwie najmniejszej liczby kolorów. Działanie algorytmu zachłannego polega na wybraniu pierwszego wierzchołka i przypisaniu mu pierwszego możliwego do przypisania koloru. W kolejnych krokach następuje przypisanie do kolejnych wierzchołków grafu kolejnego koloru, który nie jest wykorzystany dla któregokolwiek z sąsiednich wierzchołków. Tak działający algorytm, w zależności od tej kolejności wierzchołków, może rozwiązać problem w bardzo nieoptymalny sposób przez użycie do pokolorowania grafu zbyt wielu kolorów.

Pakowanie plecaka w ujęciu algorytmiki nazywamy **„problemem plecakowym”**: przy podanym zbiorze przedmiotów, z których każdy ma znaną **masę** i **wartość**, należy wybrać taki podzbiór przedmiotów, aby suma ich wartości była największa, a suma mas nie przekraczała wyznaczonej pojemności plecaka.

|Numer przedmiotu|1|2|3|4|5|6|7|
|----------------|-|-|-|-|-|-|-|
|Wartość|3|1|1|2,5|1,5|2|1,25|
|Masa|1,5|5|1|5|3|2,5|2,2|
|Opłacalność|2|0,2|1|0,5|0,5|0,8|0,1|

$opłacalność= \frac{wartość}{masa}$

8. Lista kroków:<br>
**Krok 1.** Wczytaj dane pierwszego (najbardziej opłacalnego) przedmiotu o określonej masie. <br>
**Krok 2.** Jeżeli masa przedmiotu jest równa pojemności plecaka K, wypisz wynik i zakończ algorytm. <br>
**Krok 3.** Jeżeli masa przedmiotu jest mniejsza od pojemności plecaka K, wypisz wynik. <br>
**Krok 4.** Dodaj dane kolejnego przedmiotu. <br>
**Krok 5.** Jeżeli łączna masa przedmiotów jest równa pojemności plecaka K, wypisz wynik i zakończ algorytm. <br>
**Krok 6.** Jeżeli łączna masa przedmiotów jest mniejsza od pojemności plecaka K, wypisz wynik i wykonaj dodanie danych kolejnego przedmiotu. <br>
**Krok 7.** Jeżeli suma mas przedmiotów jest równa pojemności plecaka K lub mniejsza od niej, jednak różnica jest mniejsza od masy najlżejszego z pozostałych (niespakowanych) przedmiotów, wypisz wynik i zakończ algorytm. <br>
