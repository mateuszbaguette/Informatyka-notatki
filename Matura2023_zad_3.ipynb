```python
plik = open('przyklad_pi.txt', 'r')
wiersze = plik.readlines()

wyniki3=open('wyniki3.txt','w')

# Zadanie 3.4.
max_dlugosc = 0
max_pozycja = 0
lokalna_dlugosc = 0
lokalna_pozycja = 0

for i in range(0, len(wiersze) - 5):
    czy_rosnacy = True
    czy_malejacy = True
    poprzednia_liczba = int(wiersze[i])
    j=i+1
    while j < len(wiersze) and (czy_rosnacy == True or czy_malejacy == True):
        lokalna_dlugosc+=1
        obecna_liczba=int(wiersze[j])
        if czy_rosnacy and obecna_liczba <= poprzednia_liczba:
            if lokalna_dlugosc == 1:
                lokalna_dlugosc = 0
                break
            czy_rosnacy = False
        elif not czy_rosnacy and czy_malejacy and obecna_liczba >= poprzednia_liczba:
            czy_malejacy = False
            if lokalna_dlugosc > max_dlugosc:
                max_dlugosc = lokalna_dlugosc
                max_pozycja = i
            lokalna_dlugosc = 0
        poprzednia_liczba=obecna_liczba
        j=j+1

print(max_pozycja+1,wiersze[max_pozycja:max_pozycja+max_dlugosc])
text= '\n Zad 3.4 '+ str(max_pozycja+1)+' ' + str(wiersze[max_pozycja:max_pozycja+max_dlugosc])+'\n'
print(text)
wyniki3.write(text)
```
